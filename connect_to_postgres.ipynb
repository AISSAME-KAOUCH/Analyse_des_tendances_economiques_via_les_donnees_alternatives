```python
def connect_to_db(host, dbname, port, user, password):
    try:
        conn = ps.connect(host=host, database=dbname, user=user, password=password, port=port)

    except ps.OperationalError as e:
        raise e
    else:
        print('Connected!')
        return conn
```
