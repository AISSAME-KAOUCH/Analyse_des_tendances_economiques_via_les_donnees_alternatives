```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
```

```python
data =  pd.read_excel(r'C:/Users/ADL/Desktop/Bigrammes2.xlsx')
data
```

```python
freqitemsets = apriori(data, min_support=0.001, use_colnames=True)
freqitemsets['length'] = freqitemsets['itemsets'].apply(lambda x: len(x))
```

```python
freqitemsets[:20]
```

```python
rules = association_rules(freqitemsets, metric="lift")
rules["antecedents_length"] = rules["antecedents"].apply(lambda x: len(x))
rules["consequents_length"] = rules["consequents"].apply(lambda x: len(x))
rules[1:20]
```
